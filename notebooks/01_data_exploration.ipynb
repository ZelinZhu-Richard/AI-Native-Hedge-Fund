# Data Exploration

Explore ingested OHLCV data from the Meridian database.

```python
from meridian.config.settings import get_settings
from meridian.data.storage.database import MeridianDatabase

settings = get_settings()
db = MeridianDatabase(settings.database.path, read_only=True)
db.__enter__()

# Database summary
db.get_summary()
```

```python
# Sample query: AAPL OHLCV data
df = db.get_data("AAPL")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Data quality issues
issues = db.get_tickers_with_issues()
print(f"Tickers with issues: {len(issues)}")
issues
```
